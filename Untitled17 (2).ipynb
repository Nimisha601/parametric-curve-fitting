```python
# Import basic Python tools we will use
import pandas as pd
import numpy as np
import math
from scipy.optimize import minimize

```

```python
# Load the CSV file
data = pd.read_csv("xy_data.csv")
print("First 5 rows of data:")
print(data.head())

```

```python
# Take x and y from your file
x_real = data['x'].values
y_real = data['y'].values
# Create t
t = np.linspace(6, 60, len(x_real))
print("loading done ")
print("t sample:", t[:5])
print("x sample:", x_real[:5])
print("y sample:", y_real[:5])
```

```python
#Function to calculate predicted x and y
def get_curve_points(theta, M, X, t):
    x_pred = t * np.cos(theta) - np.exp(M * np.abs(t)) * np.sin(0.3 * t) * np.sin(theta) + X
    y_pred = 42 + t * np.sin(theta) + np.exp(M * np.abs(t)) * np.sin(0.3 * t) * np.cos(theta)
    return x_pred, y_pred

```

```python
#Function for error
def find_error(params):
    theta, M, X = params
    x_pred, y_pred = get_curve_points(theta, M, X, t)
    # find how our predicted points are from real
    # we take avg of abs diff (L1 distance)
    error = np.mean(np.abs(x_real - x_pred) + np.abs(y_real - y_pred))
    return error

```

```python
# find the best values for theta, M, and X
#  start with random values
# theta (in radians), M, X
start_guess = [0.5, 0.0, 10.0]
# as given in question
bounds = [
    (0, math.radians(50)),   # theta range
    (-0.05, 0.05),           # M range
    (0, 100)                 # X range
]
result = minimize(find_error, start_guess, bounds=bounds)
theta_best, M_best, X_best = result.x
print("best Values")
print(f"Theta : {theta_best}")
print(f"theta (degree): {math.degrees(theta_best)}")
print(f"M: {M_best}")
print(f"X: {X_best}")
```

```python
import matplotlib.pyplot as plt
x_pred, y_pred = get_curve_points(theta_best, M_best, X_best, t)

plt.figure(figsize=(6,5))
plt.scatter(x_real, y_real, color='blue', label='Actual Data', alpha=0.6)
plt.plot(x_pred, y_pred, color='red', label='Predicted Curve', linewidth=2)
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Curve Fitting Result")
plt.legend()
plt.show()
```
